# Primer t-test

```python
from scipy.stats import ttest_ind
ConsumoEntreSemanaMat = alcoholEstudiantesMat['Dalc']
ConsumoEntreSemanaPor = alcoholEstudiantesPor['Dalc']

stat, p = ttest_ind(ConsumoEntreSemanaMat, ConsumoEntreSemanaPor)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

## Vamos a tomar el famoso umbral de significancia como p = 0.05 Al realizar nuestras pruebas nos vamos a enfocar en el valor P que nos da nuestros calculos, es decir al ingresar los datos a la funciÃ³n correspondiente nos enfocaremos en P 

### Como se puede ver al correr el codigo con los datos correspondientes, vemos que p nos da un valor de 0.715 esto quiere decir que nuestra p calculada es mucho mayor a nuestro umbral de significancia que habiamos establecido previamente esto quiere decir que nuestra hipotesis nula no se rechaza, es decir, podemos decir que no existe una diferencia entre la clase de matematicas y portuguÃ©s viendolo desde el punto de vista de el consumo de alcohol en dÃ­as laborales de los alumnos
